```python
import pandas as pd
```

```python
file_path = '/home/mingkang/Desktop/bootes/paper2.csv'
df = pd.read_csv(file_path)
```

```python
df.columns = ['name','ra','dec','snr','s850','s250','s350','s500','class','redshift','lir','sfr']
```

```python
df
```

```python
# for the s850 column, take out the bracketed number and put in a new column
df['err_850'] = df['s850'].str.extract(r'\((\d+\.\d+)\)',expand=False)
df['err_250'] = df['s250'].str.extract(r'\((\d+\.\d+)\)',expand=False)
df['err_350'] = df['s350'].str.extract(r'\((\d+\.\d+)\)',expand=False)
df['err_500'] = df['s500'].str.extract(r'\((\d+\.\d+)\)',expand=False)
df['s850'] = df['s850'].str.extract(r'(\d+\.\d+)',expand=False)
df['s250'] = df['s250'].str.extract(r'(\d+\.\d+)',expand=False)
df['s350'] = df['s350'].str.extract(r'(\d+\.\d+)',expand=False)
df['s500'] = df['s500'].str.extract(r'(\d+\.\d+)',expand=False)
```

```python
df
```

```python
# multiply the fluxes by 1e-3 to get them in Jy
df['s850'] = df['s850'].astype(float) * 1e-3
df['s250'] = df['s250'].astype(float) * 1e-3
df['s350'] = df['s350'].astype(float) * 1e-3
df['s500'] = df['s500'].astype(float) * 1e-3

# multiply the errors by 1e-3 to get them in Jy
df['err_850'] = df['err_850'].astype(float) * 1e-3
df['err_250'] = df['err_250'].astype(float) * 1e-3
df['err_350'] = df['err_350'].astype(float) * 1e-3
df['err_500'] = df['err_500'].astype(float) * 1e-3
```

```python
df['z_spec'] = -99
```

```python
df2 = df[['name','z_spec','s850','err_850','s250','err_250','s350','err_350','s500','err_500']]
```

```python
df2.to_csv('obervations.dat',sep='\t',index=False)
```
